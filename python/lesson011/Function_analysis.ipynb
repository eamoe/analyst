```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.Symbol('x', real=True)
```

```python
f_x = -12 * x**4 * sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
```

Для начала рассмотрим периодическую часть функции - sin(cos(x)). Эта функция будет изменять свой знак на бесконечном интервале.
Функция будет выше нуля на интервалах (-Pi/2 + Pi*n, Pi/2 + Pi*n) для любого n, принадлежащего (-inf, +inf).
Функция будет ниже нуля на интервалах (Pi/2 + 2n*Pi, 3Pi/2 + 2n*Pi) для любого n, принадлежащего (-inf, +inf).

```python
sin_cos_x = sympy.sin(sympy.cos(x))
```

```python
plot(sin_cos_x, xlim=(-10, 10), ylim=(-1, 1))
```

Теперь рассмотрим функцию f(x) = -12x^4 - 18x^3+5x^2 + 10x - 30. Эта функция очень быстро уходит в бесконечность, поэтому с учетом изложенного выше изначальная функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30 будет осциллирующей на бесконечном интервале - с ростом x она будет многократно пересекать ось ox практически вертикально, т.е. у нее будет бесконечно много корней.

```python
f_1_x = -12 * x**4 - 18 * x**3 + 5 * x**2 + 10 * x - 30
```

```python
plot(f_1_x)
```

Для простоты остановимся на интервале [-15, 15], где функция осциллирует менее выраженно и вычислим корни на этом интервале.

```python
plot(f_x, xlim=(-15, 15))
```

Будем выводить значения функции в точках, где произведение функции на текущем и следующем шаге меньше нуля. Таким образом, мы определим с некоторой точностью точку, где функция пересекает ось ox. Следующий код позволяет вычислить 10 корней, лежащих на интервале [-15, 15].

```python
delta = 0.0001
precision = 5
position = -15.0
end = 15.0
root_number = 0
while position <= end:
    if f_x.subs(x, position)*f_x.subs(x, position + delta) < 0.0:
        root_number += 1
        print(f"Root #{root_number}")
        print(f"position = {round(position, precision)}\tvalue={f_x.subs(x, position)}")
        print(f"position = {round(position + delta, precision)}\tvalue={f_x.subs(x, position  + delta)}")
    position += delta
```

Функция возрастает на интервале, где ее производная выше нуля и убывает там, где ее производная отрицательна. Вычислим производную аналитически.

```python
diff_f_x = sympy.diff(f_x)
print(diff_f_x)
```

```python
plot(diff_f_x, xlim=(-10, 10))
```

Алгоритм ниже позволяет найти точки, где производная функции пересекает ось ox. Данные точки являются точками, где изначальная функция меняет направление роста с возрастания на убывание и наоборот. Точки, в которых направление роста функции изменяется на противоположное будут вершинами функции.

```python
f_x_diff = 12*x**4*sin(x)*cos(cos(x)) - 48*x**3*sin(cos(x)) - 54*x**2 + 10*x + 10
delta = 0.001
precision = 5
position = -15.0
end = 15.0
point_number = 0
while position <= end:
    if f_x_diff.subs(x, position)*f_x_diff.subs(x, position + delta) < 0.0:
        if f_x_diff.subs(x, position) < 0.0:
            print(f"Increases from: {round(position, precision)}")
        else:
            print(f"Decreases from: {round(position, precision)}")
    position += delta
```
